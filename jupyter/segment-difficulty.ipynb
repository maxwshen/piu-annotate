```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns

df = pd.read_csv('/home/maxwshen/piu-annotate/artifacts/analysis/segment-difficulty.csv', index_col = 0)
df['pred level unpacked'] = [float(p[1:-1]) for p in df['pred level']]
df.head()
```

```python
sns.scatterplot(
    data = df,
    x = 'chart level',
    y = 'pred level unpacked',
    alpha = 0.6,
)
sns.despine()
plt.grid(alpha = 0.2)
```

```python
sns.scatterplot(
    data = df,
    x = 'pred level unpacked',
    y = 'max segment level',
    alpha = 0.6,
)
sns.despine()
plt.grid(alpha = 0.2)
```

```python
sns.scatterplot(
    data = df,
    x = 'chart level',
    y = 'max segment level',
    alpha = 0.6,
)
sns.despine()
plt.grid(alpha = 0.2)
```

```python
df['max segment level - chart level'] = df['max segment level'] - df['chart level']

df.sort_values(by = 'max segment level - chart level', ascending = False)
```

```python
df.sort_values(by = 'max segment level - chart level', ascending = False).to_csv('sorted-segment-diff.csv')
```

```python
df['adj max segment level'] = (df['chart level'] + df['max segment level']) / 2

sns.scatterplot(
    data = df,
    x = 'chart level',
    y = 'adj max segment level',
    alpha = 0.6,
)
sns.despine()
plt.grid(alpha = 0.2)
```
