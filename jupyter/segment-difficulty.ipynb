```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns

df = pd.read_csv('/home/maxwshen/piu-annotate/artifacts/analysis/segment-difficulty.csv', index_col = 0)
# df['pred level float'] = [float(p[1:-1]) for p in df['pred level']]
df.head()
```

```python
sns.scatterplot(
    data = df,
    x = 'chart level',
    y = 'max segment level',
    alpha = 0.6,
)
sns.despine()
plt.grid(alpha = 0.2)
```

```python
df['max segment level - chart level'] = df['max segment level'] - df['chart level']

df.sort_values(by = 'max segment level - chart level', ascending = False)
```

```python
sns.ecdfplot(df['max segment level - chart level'])
```

```python
sns.displot(df['max segment level - chart level'])
```

```python
df.sort_values(by = 'max segment level - chart level', ascending = False).to_csv('sorted-segment-diff.csv')
```
